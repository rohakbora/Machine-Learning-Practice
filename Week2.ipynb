```python
# Dictionary to count the number of movies each director has worked on
director_counts = {}

# Iterate over the crew column and manually extract director names
for crew_str in data1["crew"].dropna():  # Drop NaN values to avoid errors
    crew_str = crew_str.replace("'", '"')  # Standardize quotes for easier searching
    crew_list = crew_str.split("},")  # Split by "}," to separate crew members

    for crew_member in crew_list:
        if '"job": "Director"' in crew_member:  # Check if the job is "Director"
            name_start = crew_member.find('"name": "') + len('"name": "')
            name_end = crew_member.find('"', name_start)
            director_name = crew_member[name_start:name_end]  # Extract director's name

            # Count occurrences
            director_counts[director_name] = director_counts.get(director_name, 0) + 1

# Find the director with the most movies
if director_counts:
    most_frequent_director = max(director_counts, key=director_counts.get)
    print(f"The director with the most movies is: {most_frequent_director}")
else:
    print("No directors found in the dataset.")
```

```python
comedy_count = 0

for genre_str in data2["genres"].dropna():  # Drop NaN values
    genre_str = genre_str.lower()  # Convert to lowercase for case-insensitive search
    if '"name": "comedy"' in genre_str or "'name': 'comedy'" in genre_str:  # Check for Comedy genre
        comedy_count += 1

print(f"Number of movies in the 'Comedy' genre: {comedy_count}")
```
