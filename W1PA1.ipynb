```python
import pandas as pd
```

```python
csv=pd.read_csv("Preprocessing1.csv")
csv.head()
```

```python
csv.describe()
```

```python
flight_per_month=csv.groupby("Month").count()
flight_per_month
```

```python
cost_per_day=csv.groupby("WeekDay")["Price"].mean()
cost_per_day
week_day=["Monday","Tuesday","Wednesday","Thursday","Friday"]
week_end=["Saturday","Sunday"]
weekday_cost=cost_per_day.loc[week_day].mean()
print(weekday_cost)
week_end_cost=cost_per_day.loc[week_end].mean()
print(week_end_cost)
```

```python
csv["Additional_Info"] = csv["Additional_Info"].replace("No Info", "No info")
indigo_no_info_flights = csv[(csv["Airline"] == "IndiGo") & (csv["Additional_Info"] == "No info")]
len(indigo_no_info_flights)
```

```python
def hrs_to_sec(hrs):
  hrs=hrs.split()
  sec=0
  for i in hrs:
    if(i[-1:]=="h"):
      sec+=int(i[:-1])*60*60
    elif(i[-1:]=="m"):
      sec+=int(i[:-1])*60
  return sec

csv["Duration"]=csv["Duration"].apply(hrs_to_sec)
csv.describe()
```

```python
def time_to_hour_arrival(arr1):
  arr1=arr1.split()
  if(5 <= int(arr1[0][:2]) < 12):
    return "Morning"
  elif(12 <= int(arr1[0][:2]) < 17):
    return "Afternoon"
  elif(17 <= int(arr1[0][:2]) < 20):
    return "Evening"
  else:
    return "Night"

csv["Arrival_Time"]=csv["Arrival_Time"].apply(time_to_hour_arrival)
csv.head()
```

```python
def time_to_hour_dep(Dep_time):
  Dep_time=Dep_time.split()
  if(5 <= int(Dep_time[0][:2]) < 12):
    return "Morning"
  elif(12 <= int(Dep_time[0][:2]) < 17):
    return "Afternoon"
  elif(17 <= int(Dep_time[0][:2]) < 20):
    return "Evening"
  else:
    return "Night"

csv["Dep_Time"]=csv["Dep_Time"].apply(time_to_hour_arrival)
csv.head()
```

```python
#Flights that started in the morning and arrived in the evening
morn_eve = csv[(csv["Dep_Time"]=="Morning") & (csv["Arrival_Time"]=="Evening")].count()
morn_eve
```

```python
def weekday(day):
  if day in week_day:
    return 0
  elif day in week_end:
    return 1

csv["WeekDay"] = csv["WeekDay"].apply(weekday)
csv.head()
```
